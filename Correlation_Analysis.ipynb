```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
dataset=sns.load_dataset('mpg')
dataset.head()
```

```python
sns.scatterplot(data=dataset,
                x='weight',
                y='mpg')
plt.show()
```

```python
sns.scatterplot(data = dataset,
                x = "acceleration",
                y = "mpg")
plt.show()
```

```python
sns.scatterplot(data=dataset,
                x='displacement',
                y='mpg')
plt.show()
```

```python
sns.scatterplot(data=dataset,
                x='displacement',
                y='horsepower')
plt.show()
```

```python
dataset=dataset.describe()
```

```python
korelasi=dataset.corr()
korelasi
```

```python
sns.heatmap(korelasi,vmax=1,vmin=-1,center=0)
plt.show()
```

## Uji Hipotesis

```python
from scipy.stats import pearsonr
```

```python
coef,p_val=pearsonr(dataset['displacement'],dataset['mpg'])
```
