## Memisahkan Data Berdasarkan Produk(Laptop / Mobile Handphone)

```python
data_laptop=data[data['Product']=='Laptop']
data_laptop
```

```python
data_handphone=data[data['Product']=='Mobile Phone']
data_handphone
```

## Visualisasi Brand Handphone Yang Paling sering Diberi di seluruh wilayah

```python
import seaborn as sns
plt.figure(figsize=(10,5))
brand_sales = data_handphone.groupby('Brand')['Quantity Sold'].sum().sort_values(ascending=False)
sns.barplot(x=brand_sales.index, y=brand_sales.values, palette='Blues_d')
plt.title("Top Brand Handphone Yang Paling Banyak Dibeli")
plt.xticks(rotation=45)
plt.ylabel("Units Sold")
plt.xlabel("Brand")
plt.tight_layout()
plt.savefig("bar_top_brand.png", dpi=300)
plt.show()
```

## Wliayah Yang Paling Laku

```python
data_region=data_handphone['Region'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(data_region,
        labels=data_region.index,
        autopct='%1.1f%%',
        colors=['skyblue','lightcoral','lightgray','mediumseagreen','peachpuff'])
plt.title('Distrbusi Mobile Handphone')
plt.savefig("Pie_chart Handphone.png", dpi=300)
plt.tight_layout()
```

```python
data_region_west=data_handphone[data['Region']=='West']
data_region_west
```

## Brand Paling Banyak Di Beli Di Wilayah West

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
brand_sales = data_region_west.groupby(['Brand', 'Region'])['Quantity Sold'].sum().sort_values(ascending=False)
brand_sales = brand_sales.reset_index()
top_5_brands = brand_sales['Brand'].unique()[:5]
brand_sales = brand_sales[brand_sales['Brand'].isin(top_5_brands)]
sns.barplot(x='Brand', y='Quantity Sold', hue='Region', data=brand_sales, palette='Blues_d')
plt.title("Top 5 Brand Handphone Yang Paling Banyak Dibeli")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Units Sold")
plt.xlabel("Brand")
plt.tight_layout()
plt.savefig("bar_top_brand_west.png", dpi=300)
plt.show()
```

```python
data_laptop=data[data['Product']=='Laptop']
data_laptop
```
