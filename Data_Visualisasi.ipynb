###Library

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import Data

```python
dataset=sns.load_dataset('mpg')
dataset.head()
```

## Visualisasi Data

### Pie Chart

```python
data_origin=dataset['origin'].value_counts().to_dict()
data_pie=data_origin.values()
label_pie=data_origin.keys()
colors=sns.color_palette('pastel')[0:5]
plt.pie(x=data_pie,
        labels=label_pie,
        colors=colors,
        autopct='%.0f%%')
plt.title('Origin')
plt.show()
```

### Histogram

```python
plt.figure(figsize=(8,4))
sns.histplot(data=dataset,x='mpg',bins=20)
plt.show()
```

### Boxplot

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=dataset,x='mpg')
plt.show()
```

### Bar Chart

```python
plt.figure(figsize=(8,4))
sns.barplot(data=dataset,x='origin',y='mpg')
plt.show()
```

### Lineplot

```python
plt.figure(figsize=(8,4))
sns.lineplot(data=dataset,x='model_year',y='mpg',hue='origin')
```

### Scatterplot

```python
plt.figure(figsize=(8,4))
sns.scatterplot(data=dataset,x='displacement',y='mpg')
```
