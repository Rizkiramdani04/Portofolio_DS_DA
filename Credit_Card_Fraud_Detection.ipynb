```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dataset=pd.read_csv('/content/creditcard.csv')
dataset.head()
```

```python
dataset.info()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset['Class'].value_counts()
```

0 --> Normal Transaction

1 --> fraudulent transaction

```python
legit=dataset[dataset.Class==0]
fraud=dataset[dataset.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
dataset.groupby('Class').mean()
```

```python
legit_sample=legit.sample(n=492)
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
print()
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
data_train=model.predict(X_train)
data_accuracy=accuracy_score(data_train,Y_train)
print('Accuracy Training Data: ',data_accuracy)
```

```python
data_test=model.predict(X_test)
data_accuracy=accuracy_score(data_test,Y_test)
print("Accuracy Test Data: ",data_accuracy)
```
