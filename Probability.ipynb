```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Distribusi Diskret

### Distribusi Bernoulli

```python
from scipy.stats import bernoulli
cuan=bernoulli.rvs(size=1000,p=0.5)
sns.histplot(cuan)
plt.show()
```

```python
cuan
```

### Distribusi Binomial

```python
from scipy.stats import binom
pulus=binom.rvs(size=2000,n=10,p=0.5)
sns.histplot(pulus)
plt.show()
```

```python
pulus
```

### Distribusi Poisson

```python
from scipy.stats import poisson
cacat=poisson.rvs(size=1000,mu=2)
sns.histplot(cacat)
plt.show()
```

```python
cacat
```

##Distrbusi Kontinu

### Distribusi Seragam

```python
from scipy.stats import uniform
uniform=uniform.rvs(size=1000,loc=0,scale=9)
sns.histplot(uniform)
plt.show()
```

### Distribusi Normal

```python
from scipy.stats import norm
data=norm.rvs(size=1000,loc=70,scale=10)
sns.histplot(data)
plt.show()
```

### Distribusi Exponensial

```python
from scipy.stats import expon
data_tunggu=expon.rvs(scale=1,loc=5,size=1000)
sns.histplot(data_tunggu)
plt.show()
```
